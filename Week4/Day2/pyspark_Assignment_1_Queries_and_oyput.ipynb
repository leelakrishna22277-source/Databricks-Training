# **withColumnRenamed() vs alias()**

## **withColumnRenamed()**

`withColumnRenamed()` → Renames an existing column in a DataFrame.

```python
df.withColumnRenamed("emp_name", "employee_name")
```

- Works on a DataFrame.
- Renames an existing column in the DataFrame schema.
- Keeps all other columns unchanged.
- Used when you want to permanently change a column name in the DataFrame.

### Example

Before:

```text
emp_name, salary, city
```

After:

```text
employee_name, salary, city
```

---

## **alias()**

`alias()` → Renames a column while selecting it.

```python
df.select(col("emp_name").alias("employee_name"))
```

- Works on a Column.
- Used inside `select()`.
- Creates a new column name in the query result.
- Commonly used when selecting one or more columns.
- Does not modify the original DataFrame schema.

### Example

```python
df.select(
    col("emp_name").alias("employee_name"),
    col("salary").alias("monthly_salary")
)
```

Output:

```text
employee_name, monthly_salary
```

---

## **Key Difference**

| withColumnRenamed() | alias() |
|---------------------|---------|
| Works on a DataFrame | Works on a Column |
| Renames an existing column | Renames a column while selecting |
| Keeps all columns | Returns only selected columns |
| Changes DataFrame schema | Used for query output |
| Used for permanent renaming | Used for temporary/display renaming |

````python
# Rename emp_name to employee_name.
df.withColumnRenamed("emp_name","employee_name").show()
````
