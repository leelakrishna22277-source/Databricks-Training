###Read CSV files

```python
customers = spark.read.option("header", "true").option("inferSchema", "true").csv("/Volumes/workspace/default/first_volume/customers.csv")
```

```python
customers.printSchema()
customers.show()
```

#### Select specific columns

```python
display(customers.select("customer_id","first_name","city"))
```

#### Filtering

```python
display(customers.filter(customers.city=="Miami"))
```

```python
display(customers.filter(customers.first_name=="John"))
```

#### Handling Duplicates

```python
display(customers.dropDuplicates())
```

#### sorting

```python
display(customers.orderBy("first_name"))
```

## Sql Queries

```python
customers.createOrReplaceTempView("customers")
```

```python
%sql
select * from customers;
```

#### Select specific columns

```python
%sql
select customer_id, first_name, city
from customers;
```

#### Filtering

```python
%sql
select *
from customers
where state='CA';
```

```python
%sql
select *
from customers
where city='Miami';
```

```python
%sql
select *
from customers
where first_name='John';
```

#### Sorting

```python
%sql
select *
from customers
order by first_name;
```
